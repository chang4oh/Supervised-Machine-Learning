```python
# transform object to numerical 
from sklearn.preprocessing import OneHotEncoder

# splitting dataset for training and testing
from sklearn.model_selection import train_test_split

# import classification 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.svm import SVC, SVR

# import regression 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

# metrics tool (only accuracy_score is for classification)
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from math import sqrt
```

```python
#
# CLASSIFICATION 
#

# Split the dataset into predictor and target variable
X = merge_df[['Reaction Type', 'Content Type', 'Category', 'Sentiment']]
y = merge_df['Score']

# Apply OneHotEncoding
encoding = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
X_encoding = encoding.fit_transform(X)

# Split dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X_encoding, y, test_size=0.3, random_state=42)

# Dictionary of Classification Algorithms
class_models = {
    'LogReg': LogisticRegression(),
    'SVC Class': SVC(),
    'KNN Class': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
}

# train and evaluate model
# can't handle multiclass and continuous targets 
# can handle LogReg, SVM Class, KNN Class 
class_results = {}
for name, model in class_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    class_results[name] = {name:accuracy}
    
class_results
# y_test is test set for target, y_pred is the prediction to that target using X_test
```

<br>Linear Regression: line slope
<br>Decision Tree: whole tree is conditions
<br>Random Forest: multiple trees(classes) to reach single result
<br>MSE, RMSE, R^2 is metrics used to evaluate regression models 

```python
#
# CLASSIFICATION 
#

# Dictionary of Regression Algorithms
reg_models = {
    'LinReg': LinearRegression(),
    'KNN Reg': KNeighborsRegressor(),
    'SVR Reg': SVR(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
}

# train and evaluate
# apparently, all models can be placed for evaluation
# but, it makes sense for linear (continuous) predictions
reg_results = {}
for name, model in reg_models.items():
    model.fit(X_train, y_train)
    
    # predict
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # evaluation
    mse_train_reg = mean_squared_error(y_train, y_train_pred)
    mse_test_reg = mean_squared_error(y_test, y_test_pred)

    rmse_train_reg = sqrt(mse_train_reg)
    rmse_test_reg = sqrt(mse_test_reg)

    r2_train_reg = r2_score(y_train, y_train_pred)
    r2_test_reg = r2_score(y_test, y_test_pred)
    
    reg_results[name] = {'MSE train': mse_train_reg, 'MSE test': mse_test_reg, 
                         'RMSE train': rmse_train_reg, 'RMSE test': rmse_test_reg,
                         'R^2 train': r2_train_reg, 'R^2 test': r2_test_reg}

reg_results
```

<br>When test and train is at ideal MSE,RMSE -> 0 & R*2 -> 1 then it's prediction is well intended
<br>
<br>If train is good but test is bad, then it's overfit and have to do feature engineering, add regularization (caliberate fitting), or increase training data
<br>If train and test is bad, then it's underfit there need to be more data, feature, and finetune parameters
<br>Generally, adding more data and removing irrelevant column or do better data cleaning would likely solve most problems
<br>If train is bad and test is good, then it's underfit
<br>
<br>Looking above, Lin Reg is a winner with near perfection
<br>followed by KNN Reg and SVM Reg
<br>with the goal of predicting 'Score' with encoded 'Reaction Type', 'Content Type', 'Category', 'Sentiment'
